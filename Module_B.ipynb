# Interpolation Problem Statement

Given a set of data points consisting of independent and dependent variables, the goal is to find an estimation function that passes exactly through these data points. This function can then be used to predict the function values for new independent variables.

```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [1, 3, 2]

f = interp1d(x, y, kind='linear')
x_new = np.linspace(0, 2, 50)
y_new = f(x_new)

plt.plot(x, y, 'o', label='Data')
plt.plot(x_new, y_new, '-', label='Linear Interpolation')
plt.legend()
plt.show()
```

# Linear Interpolation

Given a set of data points where adjacent \(x\) values are independent variables and \(y\) values are dependent variables, the formula for estimating a value y(x) is:

y(x)=y_i+(y_{i+1}-y_i)/(x_{i+1}-x_i)*(x-x_i)

where x_i < x < x_{i+1}. Linear interpolation is accurate and efficient for estimating values within the range of the given data points.
However, it may not provide precise results for non-linear trends or extrapolation outside the data range.


```python
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4])
y = np.array([0, 1, 4, 9, 16])

interpolation_function = interp1d(x, y, kind='linear')

x_new = np.linspace(0, 4, 100)
y_new = interpolation_function(x_new)

plt.plot(x, y, 'o', label="Original Data Points")
plt.plot(x_new, y_new, '-', label="Linear Interpolation")
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Interpolation Example')
plt.show()
```

# Cubic Spline Interpolation

It provides smoothness, accuracy, and flexibility. Compared to other methods, it produces a smoother curve by constructing piecewise cubic polynomials that ensure continuity and smooth transitions at data points.

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

x = [0, 1, 2]
y = [1, 3, 2]

cs = CubicSpline(x, y, bc_type='natural')
x_new = np.linspace(0, 2, 100)
y_new = cs(x_new)

plt.plot(x, y, 'o', label='Original Data')
plt.plot(x_new, y_new, '-', label='Cubic Spline Interpolation')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Spline Interpolation')
plt.grid()
plt.show()
```
